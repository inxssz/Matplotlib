```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# simple line plot 
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
s = np.sin(2 * np.pi * x)
s
```

```python
fig, ax = plt.subplots() # OO Style

ax.plot(x, s)
ax.set(xlabel='nilai x', 
       ylabel='nilai sin(x)',
       title='Visualisasi nilai sin')
ax.grid()

plt.show()
```
